```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'group':['a', 'a', 'a','b','b','b','c','c','c'],
                    'data':[4, 3, 2, 1,12, 3, 4, 5, 7]})
data
```

### sort_values

```python
data.sort_values(by=['group', 'data'], ascending = [False, True], inplace=True)   # ascending  升序
data
# group 是降序， data 是升序    
```

```python
data.sort_values(by=['group', 'data'], ascending = [False, False], inplace=True)
data
# group 是降序， data 是升序
```

```python
data = pd.DataFrame({'food':['A1','A2','B1','B2','B3','C1','C2'],
                    'data':[1, 2, 3, 4, 5, 6, 7]})
data
```

### apply 和 map

```python
def food_map(series):
    if series['food'] == 'A1':
        return 'A'
    elif series['food'] == 'A2':
        return 'A'
    elif series['food'] == 'B1':
        return 'B'
    elif series['food'] == 'B2':
        return 'B'
    elif series['food'] == 'B3':
        return 'B'    
    elif series['food'] == 'C1':
        return 'C'   
    elif series['food'] == 'C2':
        return 'C'    

    
data['doof_map'] = data.apply(food_map, axis = 'columns')
data
```

```python
food2Upper = {
    'A1': 'A',
    'A2': 'A',
    'B1': 'B',
    'B2': 'B',
    'B3': 'B',
    'C1': 'C',
    'C2': 'C',
}

data['Upper'] = data['food'].map(food2Upper)
data
```

apply 需要指定函数   (用得更广)

map 需要指定字典

### assign

```python
data = pd.DataFrame({'data1':np.random.randn(5),
                     'data2':np.random.randn(5)})


data_ = data.assign(ration = data['data1']/data['data2'])
data_
```

```python
data_.drop('ration', axis='columns', inplace=True)
data_
```

### cut

```python
ages = [15, 18, 20, 21, 22, 34, 41, 52, 63, 79]
bins = [10, 40, 80]
bins_res = pd.cut(ages, bins)
bins_res
```

```python
bins_res.codes   # 可以看一下标签
```

```python
pd.value_counts(bins_res)
```

```python
help(pd.cut)
```

```python
group_names = ['Youth', 'Middle', 'old']
bins_res = pd.cut(ages, [10, 20, 50, 80], labels = group_names)            
pd.value_counts(bins_res)
```
