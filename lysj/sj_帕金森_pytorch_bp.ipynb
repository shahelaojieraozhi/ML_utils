```python
data = dataset.iloc[:, :9].values
iris_type = dataset.iloc[:, -2].values
print(data.shape)
print(iris_type.shape)
# (150, 4)
# (150,)
```

```python
input = torch.FloatTensor(data)
print(input.shape)
label = torch.LongTensor(iris_type)
print(label.shape)
```

```python
import torch.nn.functional as Fun
# 定义BP神经网络
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)
        self.out = torch.nn.Linear(n_hidden, n_output)
        
    def forward(self,x):
        x = Fun.relu(self.hidden(x))
        x = self.out(x)
        return x
```

```python
net = Net(n_feature=9, n_hidden=1000, n_output=3)
optimizer = torch.optim.SGD(net.parameters(), lr=0.0001)
# SGD:随机梯度下降法
loss_func = torch.nn.CrossEntropyLoss()
# 设定损失函数
```

```python
lis = []
for times in range(10):
    X_train, X_test, y_train, y_test = train_test_split(input, label, train_size=0.7)
    for i in range(10000):
        out = net(X_train)
        loss = loss_func(out, y_train)
        # 输出与label对比
        optimizer.zero_grad()
        # 初始化
        loss.backward()
        optimizer.step()
    
    
    out = net(X_test)
    # out是一个计算矩阵
    prediction = torch.max(out, 1)[1]

    # 实际y输出数据
    target_y = y_test.data.numpy()
    print(target_y)
    pred_y = prediction.numpy()
    # 预测y输出数列
    print(pred_y)
    # 计算准确率——自己写的
    L = pred_y.shape[0]
    flag = 0
    for i in range(L):
        if pred_y[i] == target_y[i]:
            flag += 1
    print("bp准确率为：", flag/L)
    lis.append(flag/L)
sum = 0
for i in lis:
    sum = sum + i
print(sum/len(lis))
```
